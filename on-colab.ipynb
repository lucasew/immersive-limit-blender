# Configuração inicial


```python
# Criação de um usuário não root pq o nix não instala at all como root
!sudo useradd -d /content -g users person
! echo 'person  ALL=(ALL) NOPASSWD:ALL' >> /etc/sudoers

# Resolve o bo do nix crashar por falta de lib
from os import environ
environ["LD_LIBRARY_PATH"] += ":/usr/lib/x86_64-linux-gnu:/lib/x86_64-linux-gnu"

# Libera a /content pra geral, não recomendável dar permissão de gravação depois que a /content/.nix-portable existir
!chmod 777 /content -R

# Curlbash do nix-portable pq o normal dá pau com o Colab, é um container malaco ¯\_(ツ)_/¯ 
!wget https://github.com/DavHau/nix-portable/releases/download/v008/nix-portable -O /tmp/nix-portable
!chmod +x /tmp/nix-portable

# Warmup do nix-portable usando um pacote bem simples, dá pra pular mas eu n vou
!su - person -c '/tmp/nix-portable nix run --verbose nixpkgs#hello'
```
